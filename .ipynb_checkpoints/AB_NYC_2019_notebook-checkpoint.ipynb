De início podemos eliminar algumas variáveis não vão interferir no preço

```python
new_dt.fillna(0, inplace=True)
```

Vamos avaliar a variável alvo e verificar seus principais indicadores estatísticos

```python
new_dt['price'].describe()
```

```python
sns.distplot(new_dt['price'], fit=norm);
fig = plt.figure()
```

```python
print("Assimetria: %f" % new_dt['price'].skew())
print("Curtose: %f" % new_dt['price'].kurt())
```

De acordo com o gráfico acima e os valores acima, podemos ver que a variável preço não pertence a uma distribuição normal, se relacionando mais com uma distrubuição de log.
Analisando o gráfico e os valores dos quartis, verificamos que a distribuição do preço possui muitos valores extremos. Necessário realizar um corte destes valores. Considerando que o valor mínimo no quarto percentil é de 175, muito distante do valor máximo de 10000, vamos utilizar um valor intermediário para realizar o corte.

```python
new_dt[new_dt['price'] > 175].count()['price']
```

Cortando exatamente em 175, vários valores são cortados. Vamos aumentar o limite

```python
new_dt[new_dt['price'] > 600].count()['price']
```

Vamos verificar a distribuição com este limite de corte

```python
threshold_600 = new_dt[new_dt['price'] < 600]
sns.distplot(threshold_600['price'], fit=norm);
fig = plt.figure()
```

Vamos diminuit o limite para 500, último pico no gráfico acima

```python
# total de valores excluídos adotante este limite
new_dt[new_dt['price'] > 500].count()['price']
```

```python
threshold_500 = new_dt[new_dt['price'] < 500]
sns.distplot(threshold_500['price'], fit=norm);
fig = plt.figure()
```

Apesar de não ser uma distribuição normal, com a adoção de um limite valores exorbitantes foram excluídos deixando a distribuição mais clara. Com o limite estabelecido, vamos seguir com a análise.

```python
df = new_dt[new_dt['price'] < 500]
```

Um histograma é muito útil para entender a distribuição das variáveis. Não vamos considerar as variáveis transformada anteriormente neste gráfico, pois elas são binárias. 
Como podemos notar, neste dataset não há muita tendência entre os dados e a única distribuição mais parecida com uma distribuição normal foram latitude e longitude. 

```python
hist_columns = ['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
df[hist_columns].hist(figsize=(30,30))
```

```python
# Para verificar a relação entre as variáveis utilizamos uma matriz de correlação. Novamente, as variáveis não
# apresentam uma correlação importante entre si
corrmat = df.corr()
f, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(corrmat, vmax=.8, square=True, annot=True);
```

De acordo com a matriz de correlação, não foram encontradas variáveis com forte ou fraca correlação com o preço. Possivelmente estes resultados se devem ao fato de que as variáveis mais relacionadas ao preço sejam categóricas.
